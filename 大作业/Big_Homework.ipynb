```python
import numpy as np
import matplotlib.pyplot as plt
import Molecular_Dynamics as MD
import MonteCarlo as MC
import Optimize as Optimize
import scipy.constants as constants
```

1. 二维伊辛模型的蒙特卡洛模拟

```python
Ising = MC.Ising2D(
    1, constants.k, 0, configuration="cold", size=(40, 40)
)  # *根据计算结果可以看出这个温度低于临界温度
Ising.Calculate_info()
Ising.Print_Info()
Ising.Configuration_Image()
Ising.Evolution(1000)
Ising.Print_Info()
Ising.Configuration_Image()
Ising.Initialize()
Ising.T = 10  # *高于临界温度
Ising.Evolution(1000)
Ising.Print_Info()
Ising.Configuration_Image()
Ising.Initialize()
Ising.T = Ising.T_c  # *将温度设置为临界温度，研究慢化现象
Ising.Evolution(10)
Ising.Print_Info()
Ising.Configuration_Image()
Ising.Evolution(100)
Ising.Print_Info()
Ising.Configuration_Image()
Ising.Evolution(1000)
Ising.Print_Info()
Ising.Configuration_Image()
```

从上面的程序运行结果中可以看到，当温度小于临界温度时，不会发生自发磁化；而当温度大于临界温度时会发生自发磁化。当温度在临界温度附近时，虽然也会有自发磁化现象，但是状态随时间演化得相当缓慢

2. 分子动力学的FPU模拟

```python
FPU1 = MD.FPU(6, 1, 1, 0, 0)
FPU1.Evolution(step=1000)
FPU1.Show_Dynamic()
FPU1.Phase_Diagram()
FPU1.Evolution(step=100000)
FPU1.Phase_Diagram()
```

从以上模拟结果中可以看到，经过很长时间后，粒子将会遍历相空间中大部分区域。可以推测，当经过无穷长时间后，粒子将会遍历相空间中所有区域
